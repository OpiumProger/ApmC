```python
from rdkit import Chem
from pathlib import Path

sdf_dir = Path("ligands_sdf")
sdf_dir.mkdir(exist_ok=True)

supplier = Chem.SDMolSupplier("ligands_3d.sdf", removeHs=False)

count = 0
for i, mol in enumerate(supplier):
    if mol is None:
        continue
    name = mol.GetProp("_Name") if mol.HasProp("_Name") else f"lig_{i:04d}"
    writer = Chem.SDWriter(str(sdf_dir / f"{name}.sdf"))
    writer.write(mol)
    writer.close()
    count += 1

print("Prepared:", count)
```

```python
from rdkit import Chem
from collections import defaultdict

in_sdf = "ligands_3d.sdf"
tmp_sdf = "ligands_3d_tmp.sdf"

supplier = Chem.SDMolSupplier(in_sdf, removeHs=False)
writer = Chem.SDWriter(tmp_sdf)

counts = defaultdict(int)

for mol in supplier:
    if mol is None:
        continue

    raw_name = mol.GetProp("_Name") if mol.HasProp("_Name") else ""
    source = raw_name.split("_i")[0]  # 1_i0 → 1

    if counts[source] < 2:  # 🔥 ограничение
        writer.write(mol)
        counts[source] += 1

writer.close()

# 🔥 ВАЖНО: перезаписываем исходный файл
import os
os.replace(tmp_sdf, in_sdf)

print("✅ Ограничили до max 2 структур на молекулу")
```

```python
from pathlib import Path
import subprocess

sdf_dir = Path("ligands_sdf")
pdbqt_dir = Path("ligands_pdbqt")
pdbqt_dir.mkdir(exist_ok=True)

failed = []

for sdf_file in sdf_dir.glob("*.sdf"):
    out_file = pdbqt_dir / f"{sdf_file.stem}.pdbqt"

    cmd = [
        "obabel",
        str(sdf_file),
        "-O", str(out_file),
        "-h",
        "--partialcharge", "gasteiger"
    ]

    result = subprocess.run(cmd, capture_output=True, text=True)
    if result.returncode != 0 or not out_file.exists():
        failed.append((sdf_file.name, result.stderr[:300]))

print("Prepared PDBQT:", len(list(pdbqt_dir.glob('*.pdbqt'))))
print("Failed:", len(failed))
failed[:5]
```

```python
RECEPTOR = "/content/1XGIprot.pdbqt"   # замени на имя своего файла

CENTER_X = 22.12
CENTER_Y = 8.08
CENTER_Z = 10.79

SIZE_X = 15.0
SIZE_Y = 15.0
SIZE_Z = 15.0

EXHAUSTIVENESS = 8
NUM_MODES = 9
CPU = 2
```

```python
import subprocess
from pathlib import Path
import pandas as pd
import re

dock_dir = Path("qvina_out")
dock_dir.mkdir(exist_ok=True)

results = []

def parse_best_affinity(log_text):
    # Ищем первую строку таблицы поз, например:
    # 1   -8.2   0.000   0.000
    for line in log_text.splitlines():
        m = re.match(r"^\s*1\s+(-?\d+\.\d+)\s+", line)
        if m:
            return float(m.group(1))
    return None

for lig_file in pdbqt_dir.glob("*.pdbqt"):
    lig_name = lig_file.stem
    out_pdbqt = dock_dir / f"{lig_name}_out.pdbqt"
    log_file = dock_dir / f"{lig_name}.log"

    cmd = [
        "/content/qvina/qvina02",
        "--receptor", RECEPTOR,
        "--ligand", str(lig_file),
        "--center_x", str(CENTER_X),
        "--center_y", str(CENTER_Y),
        "--center_z", str(CENTER_Z),
        "--size_x", str(SIZE_X),
        "--size_y", str(SIZE_Y),
        "--size_z", str(SIZE_Z),
        "--exhaustiveness", str(EXHAUSTIVENESS),
        "--num_modes", str(NUM_MODES),
        "--cpu", str(CPU),
        "--out", str(out_pdbqt)
    ]

    result = subprocess.run(cmd, capture_output=True, text=True)
    full_log = (result.stdout or "") + "\n" + (result.stderr or "")
    log_file.write_text(full_log)

    best_aff = parse_best_affinity(full_log)

    results.append({
        "name": lig_name,
        "best_affinity_kcal_mol": best_aff,
        "pose_file": str(out_pdbqt) if out_pdbqt.exists() else None,
        "status": "ok" if result.returncode == 0 else "failed"
    })

res_df = pd.DataFrame(results).sort_values("best_affinity_kcal_mol", ascending=True)
res_df.to_csv("qvina_results.csv", index=False)
res_df.head(20)
```

```python
# --- 1. обработка результатов докинга ---
res_df = res_df.copy()
res_df["name"] = res_df["name"].astype(str)

# вытаскиваем номер исходной строки: 1_i0 -> 1
res_df["source_row"] = res_df["name"].str.extract(r"^(\d+)")

# оставляем только строки, где affinity посчиталась
res_valid = res_df.dropna(subset=["best_affinity_kcal_mol"]).copy()

# находим лучший вариант для каждой исходной молекулы
# для docking лучше = более отрицательная affinity
best_idx = res_valid.groupby("source_row")["best_affinity_kcal_mol"].idxmin()

# сохраняем не только affinity, но и pose_file, и name победившего варианта
res_best = res_valid.loc[
    best_idx,
    ["source_row", "name", "pose_file", "best_affinity_kcal_mol"]
].copy()

# --- 2. возвращаем в исходный top15 ---
top15 = top15.copy().reset_index(drop=True)

# ключ для merge: индекс строки top15
top15["source_row"] = top15.index.astype(str)

df_final = top15.merge(
    res_best,
    on="source_row",
    how="left"
)

# если source_row не нужен — убираем
df_final = df_final.drop(columns=["source_row"])

df_final
```

```python
import py3Dmol
from pathlib import Path
import pandas as pd

def show_one_hit(i, table=df_final, receptor_path="vis_pdb/1XGIprot.pdb",
                 affinity_col="best_affinity_kcal_mol", smiles_col="SMILES",
                 pic50_col=None):
    row = table.iloc[i]

    lig_name = str(row["name"])
    lig_path = Path(f"vis_pdb/{lig_name}.pdb")

    if not lig_path.exists():
        print(f"Нет файла позы для {lig_name}: {lig_path}")
        return

    affinity = row[affinity_col] if affinity_col in table.columns else None
    smiles = row[smiles_col] if smiles_col in table.columns else None
    pic50 = row[pic50_col] if (pic50_col is not None and pic50_col in table.columns) else None

    title = f"#{i} | name={lig_name}"
    if pd.notna(affinity):
        title += f" | affinity={affinity:.2f} kcal/mol"
    if pic50 is not None and pd.notna(pic50):
        title += f" | pIC50={pic50:.2f}"

    print(title)
    if smiles is not None:
        print("SMILES:", smiles)

    view = py3Dmol.view(width=900, height=650)

    with open(receptor_path, "r") as f:
        receptor_block = f.read()
    with open(lig_path, "r") as f:
        ligand_block = f.read()

    view.addModel(receptor_block, "pdb")
    view.setStyle({"model": 0}, {"cartoon": {"color": "lightgray"}})
    view.addSurface(py3Dmol.VDW, {"opacity": 0.18}, {"model": 0})

    view.addModel(ligand_block, "pdb")
    view.setStyle({"model": 1}, {"stick": {}})

    view.zoomTo({"model": 1})
    view.show()
```

```python
import ipywidgets as widgets
from IPython.display import display

current_idx = 0

out = widgets.Output()

def show(i):
    out.clear_output(wait=True)
    with out:
        show_one_hit(i)  # твоя функция визуализации

def on_prev(b):
    global current_idx
    if current_idx > 0:
        current_idx -= 1
    show(current_idx)

def on_next(b):
    global current_idx
    if current_idx < len(df_final) - 1:
        current_idx += 1
    show(current_idx)

btn_prev = widgets.Button(description="⬅️ Prev")
btn_next = widgets.Button(description="Next ➡️")

btn_prev.on_click(on_prev)
btn_next.on_click(on_next)

display(widgets.HBox([btn_prev, btn_next]))
display(out)

# показать первую молекулу
show(current_idx)
```
